# Plotting 

```python
house_df.plot(kind  = 'scatter', x = 'condition' , y = 'price')
plt.show()
```

```python
sns.scatterplot(x = 'bedrooms', y= 'bathrooms' , data = house_df , hue='floors')
```

```python
sns.scatterplot(x = 'yr_built', y= 'yr_renovated' , data = house_df , hue='bedrooms')
```

```python
sns.scatterplot(x = 'price', y= 'view' , data = house_df , hue='date')
```

```python
house_df.plot()
```

```python
sns.pairplot(house_df , diag_kind = 'hist')
```

```python
h= sns.FacetGrid(house_df , hue = 'city')
h.map(plt.hist, 'yr_built')
```

```python
h= sns.FacetGrid(house_df , hue = 'city')
h.map(sns.distplot, 'condition')
```

```python
from pandas.plotting import boxplot , scatter_matrix 
```

```python
scatter_matrix(house_df)
plt.plot()
```

```python
x= house_df.price
y =house_df.bedrooms
plt.scatter(x,y,c='b')
plt.title("Scatter Graph")
plt.ylabel('bedrooms')
plt.xlabel('price')
plt.title('bedrooms vs price')
plt.show()
```
